# Data preparation

Plan:
Do a function for flat, climb, time trial)
def (time_trial, profilescore_min, profile_score_max, col_name)
- nb_race_col_name: count
- nb_win_col_name: count with filter rnk_finished=1
- nb_pod_col_name: count with filter rnk_finished=1 or 2 or 3
- nb_top10_col_name: count with filter rnk_finished=1 to 10
- pts_col_name: sum of column 'pnt'
- avg_rnk_col_name = average of rnk_finished
- pct_win_col_name = nb_win_col_name / nb_race_col_name
- pct_pod_col_name = nb_pod_col_name / nb_race_col_name
- pc_top10_col_name = nb_top10_col_name / nb_race_col_name
- avg_pts_col_name = pts_col_name / nb_race_col_name
where col_name is 'flat', 'climb', 'time_trial'
flat : time_trial=0, 0<=profilescore<=50
climb : time_trial=0, 51<=profilescore<=9999
time_trial : time_trial=1, 0<=profilescore<=9999

for one-day-race, will reuse the function but a type_race=one-day-race filter ; time_trial=0, 0<=profilescore<=9999

for GC, we will have to adapt more in details the code
- number of gc finished = count of (final_stage_x=1)
- number win gc = count of (final_stage_x=1, gc=1)
- number podium gc = count of (final_stage_x=1, gc=1 or 2 or 3)
- number top 10 gc = count of (final_stage_x=1, gc=1 to 10)
- total pcs points gc (gc_pcs_point)
- avg rnk gc = average of gc (final_stage_x=1)
- % win gc = number win gc  / number of gc finished
- % podium gc = number podium gc / number of gc finished
- % top 10 gc = number top 10 gc / number of gc finished
- avg pcs points gc

and a final one: 
- number of dnf/otl
- % of dnf/otl

# Creating the summary performance data per riders

```python
# creating a function for getting all the results type of race by type of race
def performance_calculus(df, time_trial, profilescore_min, profilescore_max, col_name):
    df_copy = df.copy()

    df_copy = df_copy[(df_copy['profilescore'] >= profilescore_min) & (df_copy['profilescore'] <= profilescore_max)]

    if time_trial == 0:
        df_copy = df_copy[df_copy['time_trial'] == 0]
    elif time_trial == 1:
        df_copy = df_copy[df_copy['time_trial'] == 1]

    result_df = pd.DataFrame(columns=['rider_code', 'nb_race_'+col_name, 'nb_win_'+col_name, 'nb_pod_'+col_name,
                                       'nb_top10_'+col_name, 'pts_'+col_name, 'avg_rnk_'+col_name,
                                       'pct_win_'+col_name, 'pct_pod_'+col_name, 'pct_top10_'+col_name,
                                       'avg_pts_'+col_name])

    rider_codes = df_copy['rider_code'].unique()
    for rider_code in rider_codes:
        rider_df = df_copy[df_copy['rider_code'] == rider_code]

        nb_race = len(rider_df)
        if nb_race != 0:
            nb_win = len(rider_df[rider_df['rnk_finished'] == 1])
            nb_pod = len(rider_df[rider_df['rnk_finished'].isin([1,2,3])])
            nb_top10 = len(rider_df[rider_df['rnk_finished'].isin(range(1,11))])
            pts = rider_df['pnt'].sum()
            avg_rnk = rider_df['rnk_finished'].mean()
            pct_win = nb_win / nb_race
            pct_pod = nb_pod / nb_race
            pct_top10 = nb_top10 / nb_race
            avg_pts = pts / nb_race
        elif nb_race == 0:
            nb_win = None
            nb_pod = None
            nb_top10 = None
            pts = None
            avg_rnk = None
            pct_win = None
            pct_pod = None
            pct_top10 = None
            avg_pts = None
        
        result_df = result_df.append({'rider_code': rider_code,
                                      'nb_race_'+col_name: nb_race,
                                      'nb_win_'+col_name: nb_win,
                                      'nb_pod_'+col_name: nb_pod,
                                      'nb_top10_'+col_name: nb_top10,
                                      'pts_'+col_name: pts,
                                      'avg_rnk_'+col_name: avg_rnk,
                                      'pct_win_'+col_name: pct_win,
                                      'pct_pod_'+col_name: pct_pod,
                                      'pct_top10_'+col_name: pct_top10,
                                      'avg_pts_'+col_name: avg_pts}, ignore_index=True)

    return result_df
```

```python
def performance_calculus_dnf_otl(df):
    df_copy = df.copy()

    result_df = pd.DataFrame(columns=['rider_code','nb_dnf_otl', 'pct_dnf_otl'])

    rider_codes = df_copy['rider_code'].unique()
    for rider_code in rider_codes:
        rider_df = df_copy[df_copy['rider_code'] == rider_code]

        nb_race = len(rider_df)
        if nb_race != 0:
            nb_dnf_otl = len(rider_df[(rider_df['rnk'] == 'dnf') | (rider_df['rnk']== 'otl')])
            pct_dnf_otl = nb_dnf_otl / nb_race

        elif nb_race == 0:
            nb_dnf_otl = None
            pct_dnf_otl = None
        
        result_df = result_df.append({'rider_code': rider_code,
                                      'nb_dnf_otl': nb_dnf_otl,
                                      'pct_dnf_otl': pct_dnf_otl}, ignore_index=True)

    return result_df
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
pairs = [('pct_win_flat', 'pct_win_climb'),
                ('pct_pod_flat', 'pct_pod_climb'),
                ('pct_top10_flat', 'pct_top10_climb'),
                ('pts_flat', 'pts_climb'),
                ('avg_pts_flat', 'avg_pts_climb'),
                ('avg_rnk_flat', 'avg_rnk_climb')]

for i, ax in enumerate(axs.flatten()):
    x_col, y_col = pairs[i]
    ax.scatter(rider_df_all_perf[x_col], rider_df_all_perf[y_col], alpha=0.5)

corr_1 = rider_df_all_perf['pct_win_flat'].corr(rider_df_all_perf['pct_win_climb'])
corr_2 = rider_df_all_perf['pct_pod_flat'].corr(rider_df_all_perf['pct_pod_climb'])
corr_3 = rider_df_all_perf['pct_top10_flat'].corr(rider_df_all_perf['pct_top10_climb'])
corr_4 = rider_df_all_perf['pts_flat'].corr(rider_df_all_perf['pts_climb'])
corr_5 = rider_df_all_perf['avg_pts_flat'].corr(rider_df_all_perf['avg_pts_climb'])
corr_6 = rider_df_all_perf['avg_rnk_flat'].corr(rider_df_all_perf['avg_rnk_climb'])

axs[0, 0].set_title('Correlation between % win flat & climb, corrr = {:.2f}'.format(corr_1))
axs[0, 0].set_xlabel('% win flat')
axs[0, 0].set_ylabel('% win climb')
axs[0, 1].set_title('Correlation between % podium flat & climb, corr = {:.2f}'.format(corr_2))
axs[0, 1].set_xlabel('% podium flat')
axs[0, 1].set_ylabel('% podium climb')
axs[0, 2].set_title('Correlation between % top 10 flat & climb, corr = {:.2f}'.format(corr_3))
axs[0, 2].set_xlabel('% top 10 flat')
axs[0, 2].set_ylabel('% top 10 climb')
axs[1, 0].set_title('Correlation between points flat & climb, corr = {:.2f}'.format(corr_4))
axs[1, 0].set_xlabel('points in flat')
axs[1, 0].set_ylabel('points in climb')
axs[1, 1].set_title('Correlation between average points flat & climb, corr = {:.2f}'.format(corr_5))
axs[1, 1].set_xlabel('average points in flat')
axs[1, 1].set_ylabel('average points in climb')
axs[1, 2].set_title('Correlation between average rank flat & climb, corr = {:.2f}'.format(corr_6))
axs[1, 2].set_xlabel('average rank in flat')
axs[1, 2].set_ylabel('average rank in climb')

fig.suptitle('Performance comparison between flat and climb races')
fig.tight_layout()
plt.show()
```

```python
team_df_wt = team_df_all_perf.loc[team_df_all_perf['circuit'] == 'worldteams']

team_df_wt1 = team_df_wt[['team_name', 'nb_win_flat', 'nb_win_climb', 'nb_win_time_trial']]
team_df_wt1['total_wins'] = team_df_wt1[['nb_win_flat', 'nb_win_climb', 'nb_win_time_trial']].sum(axis=1)
team_df_wt1 = team_df_wt1.sort_values('total_wins', ascending=True)
team_df_wt1.drop('total_wins', axis=1, inplace=True)

team_df_wt2 = team_df_wt[['team_name', 'nb_win_one_day', 'nb_win_gc']]
team_df_wt2['total_wins'] = team_df_wt2[['nb_win_one_day', 'nb_win_gc']].sum(axis=1)
team_df_wt2 = team_df_wt2.sort_values('total_wins', ascending=True)
team_df_wt2.drop('total_wins', axis=1, inplace=True)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 8))

colors_1=['green','orange','black']
colors_2=['blue','yellow']

team_df_wt1.plot(kind='barh', stacked=True, color=colors_1, ax=ax1)
ax1.set_title('Team wins by type of races')
ax1.set_xlabel('Number of wins')
ax1.set_yticks(range(len(team_df_wt1)))
ax1.set_yticklabels(team_df_wt1['team_name'])
ax1.legend(loc='lower right')

team_df_wt2.plot(kind='barh', stacked=True, color=colors_2, ax=ax2)
ax2.set_title('Team wins for one-day race vs stage race')
ax2.set_xlabel('Number of wins')
ax2.set_yticks(range(len(team_df_wt2)))
ax2.set_yticklabels(team_df_wt2['team_name'])
ax2.legend(loc='lower right')

plt.tight_layout()
plt.show()
```
