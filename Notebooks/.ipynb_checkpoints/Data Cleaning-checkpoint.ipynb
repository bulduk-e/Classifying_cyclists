```python
# uploading the df
def opening_df(folder_path):
    all_dfs = {}
    columns_df = pd.DataFrame()

    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            file_path = os.path.join(folder_path, file)
            file_name = file.split('.')[0]

            df = pd.read_csv(file_path)
            all_dfs[file_name] = df

            # Create a DataFrame with column names for each file
            columns_df[file_name] = pd.Series(df.columns)

    return columns_df
```

```python
# Splitting race code into 3 new columns
def split_race_code(df, column_name):
        
    df_parts = df[column_name].str.split('/', expand=True)
    
    if len(df_parts.columns) == 3:
        df_parts.columns = ['race_name', 'year', 'stage_number']
    elif len(df_parts.columns) == 2:
        df_parts.columns = ['race_name', 'year']
        df_parts['stage_number'] = ''
    else:
        raise ValueError('Column must contain one or two "/" characters')
    
    if 'stage_number' in df_parts.columns:
        df_parts['stage_number'] = df_parts['stage_number'].str.replace('stage-', '')
    
    df_new = pd.concat([df, df_parts], axis=1)
        
    return df_new
```

```python
# Add a column to identify final stage for stage race
def identify_final_stage(df, race_name='race_name', year='year', 
                         stage_number='stage_number'):
    # Convert stage_number to integer type as it is currently object type
    df[stage_number] = df[stage_number].astype(int)

    max_stage = df.groupby([race_name, year])[stage_number].transform(max)

    df['final_stage'] = df[stage_number] == max_stage

    df['final_stage'] = df['final_stage'].map({True: 'True', 
                                               False: 'False'})
    
    return df
```

```python
# For date transform datetime, keep only local time ; for avg speed, remove "km/h" ; for distance remove "km" ; avg temperature remove 'Â°c'
def clean_date_time_km(df):
    df['date'] = pd.to_datetime(df['date'], format='%d %B %Y')    
    df['avg_speed_winner'] = df['avg_speed_winner'].str.replace(' km/h', '').astype(float)
    df['distance'] = df['distance'].str.replace(' km', '').astype(float)
    return df
```

```python
# somehow df_race_info_stage_2019 has a different format
def clean_date_time_km_other_type(df):
    df['date'] = pd.to_datetime(df['date'], format='%d-%b-%y')    
    df['avg_speed_winner'] = df['avg_speed_winner'].str.replace(' km/h', '').astype(float)
    df['distance'] = df['distance'].str.replace(' km', '').astype(float)
    return df
```

```python
def format_time(df, column_name):

    def extract_time(x):
        return x.split(' ')[0] if ' ' in x else x
    
    df[column_name] = df[column_name].apply(extract_time)
    
    df[column_name] = pd.to_datetime(df[column_name], format='%H:%M')
    
    df[column_name] = df[column_name].dt.strftime('%H:%M')
```

```python
def assign_points(group):
    if group['points_scale_x'].iloc[0] == 'gt.a':
        points_list = gt_a
    elif group['points_scale_x'].iloc[0] == 'gt.b':
        points_list = gt_b
    elif group['points_scale_x'].iloc[0] == '2.wt':
        points_list = two_wt
    else:
        raise ValueError('Invalid points_scale value')
    group = group.sort_values(by='gc')
    group['gc_pcs_point'] = points_list[:len(group)]
    return group

# Group by race_code and apply the function to each group
df_final_stage_finished = df_final_stage_finished.reset_index(drop=True)
df_final_stage_finished = df_final_stage_finished.reset_index(drop=True).groupby('race_code')
df_final_stage_finished = df_final_stage_finished.apply(assign_points).reset_index(drop=True)
```

```python
df_final_stage_finished = df_final_stage_finished.reset_index(drop=True).groupby('race_code').apply(assign_points).reset_index(drop=True)
```
